```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
path = "https://raw.githubusercontent.com/aztgmb/Recommender_System/master/new_data_merged(1).csv"
data = pd.read_csv(path)
data = data.drop(data.columns[0],axis=1)
```

```
graphs = data[['userID', 'itemID', 'overall', 'price']]
```

```
pd.value_counts(graphs['overall']).plot.bar()
```

```
plt.figure(figsize=(15,5))
plt.scatter(graphs['userID'], graphs['overall'])
```

```
plt.figure(figsize=(15,5))
plt.scatter(graphs['itemID'], graphs['overall'])
```
